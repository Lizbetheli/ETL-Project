```python
# Dependencies
import os
from bs4 import BeautifulSoup 
import requests
```

```python
url = "https://nces.ed.gov/ccd/tables/ACGR_RE_and_characteristics_2015-16.asp"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table = soup.find_all('table')
print(table)
```

```python
#  table_data = table.find_all("tr") 
import pandas as pd

rows = table[1].find('tbody').find_all('tr')
all_rows=[]
#row=[]
for each_row in rows[:5]: 
    row=[]
    for each_cell in each_row.find_all('td'): 
        row.append(each_cell.text)
    all_rows.append(row)
pd.DataFrame(all_rows).head()
```
